# Importing the libraries

```python

import numpy as np
import pandas as pd
import requests 
from bs4 import BeautifulSoup 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

# Obtaining and cleaning the data

```python
wikiurl="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
soup = BeautifulSoup(response.text, 'html.parser')
canadatable=soup.find('table',{'class':"wikitable sortable"})
df=pd.read_html(str(canadatable))
df=pd.DataFrame(df[0])
df=df.replace("Not assigned",np.nan)
df.dropna(inplace=True)
canada=pd.get_dummies(df)
canada.head(20)




```

```python
km=KMeans(n_clusters=3)
pred=km.fit_predict(canada)
canada['Cluster']=pred
df1=canada[canada.Cluster==0]
df2=canada[canada.Cluster==1]
df3=canada[canada.Cluster==1]
plt.scatter(df1,color='red')
plt.scatter(df1,color='green')
plt.scatter(df1,color='blue')
```
